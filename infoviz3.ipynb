```python
app = dash.Dash(__name__)
```

```python
fig1 = px.scatter(df_subset, x='Deaths per year', y='Percent who die', color='Pathogen_type',symbol='Pathogen_type',
                  size_max=10, log_x=True, width=800, height=800,
                  hover_data=['Disease name', 'Contagiousness', 'Deaths per year', 'Percent who die',
                              'New cases per year', 'Number of sufferers']
                  )

# Create the scatter plot for 'fatal vs fatal'
fig2 = px.scatter(df_subset, x='Deaths per year', y='Deaths per year', color='Pathogen_type',symbol='Pathogen_type',
                  size_max=10, log_x=True,log_y=True, width=1000, height=1000,
                  hover_data=['Disease name', 'Contagiousness', 'Deaths per year', 'Percent who die',
                              'New cases per year', 'Number of sufferers']
                  )

fig3 = px.scatter(df_subset, x='Percent who die', y='Percent who die', color='Pathogen_type',symbol='Pathogen_type',
                  size_max=10, width=800, height=800,
                  hover_data=['Disease name', 'Contagiousness', 'Deaths per year', 'Percent who die',
                              'New cases per year', 'Number of sufferers']
                  )

fig4 = px.scatter(df_subset, x='Percent who die', y='Deaths per year', color='Pathogen_type',symbol='Pathogen_type',
                  size_max=10, log_y=True, width=800, height=800,
                  hover_data=['Disease name', 'Contagiousness', 'Deaths per year', 'Percent who die',
                              'New cases per year', 'Number of sufferers']
                  )

```

```python
# Create the layout of the app
app.layout = html.Div([
    dcc.Dropdown(
        id='graph-dropdown',
        options=[
            {'label': 'Deadliness vs Fatal', 'value': 'deadliness_vs_fatal'},
            {'label': 'Deadliness vs Deadliness', 'value': 'deadliness_vs_deadliness'},
            {'label': 'Fatal vs Fatal', 'value': 'fatal_vs_fatal'},
            {'label': 'Fatal vs Deadliness','value': 'fatal_vs_deadliness'}
        ],
        value='deadliness_vs_fatal',  # Default selected value
    ),
    dcc.Graph(id='scatter-plot')
])

```

```python
# Define a callback to update the scatter plot based on the selected dropdown value
@app.callback(
    Output('scatter-plot', 'figure'),
    [Input('graph-dropdown', 'value')]
)
def update_scatter_plot(selected_value):
    if selected_value == 'deadliness_vs_fatal':
        return fig1
    elif selected_value == 'fatal_vs_fatal':
        return fig2
    elif selected_value == 'deadliness_vs_deadliness':
      return fig3
    elif selected_value == 'fatal_vs_deadliness':
      return fig4

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
